```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] = 14
plt.rcParams['font.weight'] = 'bold'
plt.style.use('seaborn-whitegrid')
```

```python
path = '../Downloads/'
df = pd.read_csv(path + 'Hotel Reservations.csv')
print('\n Number of rows and columns in the data set: ', df.shape)
print('')
df.head()
```

```python
sns.lmplot(x='lead_time', y='avg_price_per_room', data=df, aspect=2, height=6)
plt.xlabel('Tiempo de espera de reserva') 
plt.ylabel('Precio por habitación')
plt.title('Tiempo Vs Precio')
```
